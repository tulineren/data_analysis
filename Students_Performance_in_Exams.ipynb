```r
library(tidyverse) 
df <- read.csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
#Cinsiyete gore ortalama puan farki
df%>%
  group_by(gender)%>% 
  summarise(ortalama_math = mean(math.score))

#Hazirlik kursuna katilanlarin basari durumu
new_data <- df%>%
  filter(test.preparation.course == "completed")%>%
  select(math.score, reading.score, writing.score)%>%
  summarise(
      ortalama_math = mean(math.score),
      ortalama_reading = mean(reading.score),
      ortalama_writing = mean(writing.score)
  )
new_data

#Ebeveyn egitim seviyesinin etkisi
df2 <- df%>%
  group_by(parental.level.of.education)%>%
  summarise(
      ortalama_math = mean(math.score),
      ortalama_reading = mean(reading.score),
      ortalama_writing = mean(writing.score),
      kisi_sayisi = n() #ogrenci sayisini gosterir
  )%>%
  arrange(desc(ortalama_math))
df2

#Matematik notu icin gorsells≈ütirme (ggplot2 ile barplot)
ggplot(df, aes(x = parental.level.of.education, y = math.score))+
  geom_boxplot(fill = "lightblue")+
  labs(
      title = "Ebeveyn Egitim Seviyesine Gore Matematik Notlari",
      x = "Ebeveyn Egitim Seviyesi",
      y = "Matematik Notu"
  )
#theme(axis.text.x = element_text(angle = 45, hjust = 1))







```
